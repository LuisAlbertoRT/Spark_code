```python


# 1. Inicializar la Sesión de Spark (El punto de entrada a PySpark)
spark = SparkSession.builder \
    .appName("AnalisisTransaccionesCompliance") \
    .getOrCreate()

print("¡Sesión de Spark creada con éxito!")
```

```python


# 2. Crear un dataset sintético (Simulando millones de registros)
# En la vida real, aquí usarías: spark.read.csv("ruta/archivo.csv", header=True)
data = [
    ("TX1001", "Cliente_A", 15000.00, "Transferencia", "CDMX"),
    ("TX1002", "Cliente_B", 2500.50, "Retiro", "Guadalajara"),
    ("TX1003", "Cliente_A", 95000.00, "Depósito", "CDMX"),
    ("TX1004", "Cliente_C", 450.00, "Compra", "Monterrey"),
    ("TX1005", "Cliente_B", 120000.00, "Transferencia", "Guadalajara"),
    ("TX1006", "Cliente_A", 8000.00, "Retiro", "CDMX"),
    ("TX1007", "Cliente_D", 3500.00, "Compra", "Cancún")
]

columnas = ["ID_Transaccion", "ID_Cliente", "Monto", "Tipo", "Ciudad"]

# Convertimos la lista en un DataFrame de PySpark (distribuido)
df = spark.createDataFrame(data, schema=columnas)

print("Mostrando el DataFrame:")
df.show()
```

```python

# 3. Transformación de Datos: Filtrar transacciones grandes (> $10,000)
print("\n--- Transacciones mayores a $10,000 (Potencial alerta de Compliance) ---")
df_altos_montos = df.filter(col("Monto") > 10000)
df_altos_montos.show()
```

```python

# 4. Agregación: Calcular el total gastado y promedio por cliente
print("\n--- Resumen financiero por Cliente ---")
perfil_cliente = df.groupBy("ID_Cliente").agg(
    sum("Monto").alias("Total_Movilizado"),
    avg("Monto").alias("Promedio_Por_TX"),
    count("ID_Transaccion").alias("Numero_Transacciones")
)
perfil_cliente.show()
```

```python

# 5. Agregar una columna condicional (Etiquetado de Riesgo)
print("\n--- Clasificación de Riesgo del Cliente ---")
# Si el total movilizado supera los 100,000, es riesgo Alto.
perfil_con_riesgo = perfil_cliente.withColumn(
    "Nivel_Riesgo",
    when(col("Total_Movilizado") >= 100000, "ALTO").otherwise("ESTÁNDAR")
)
perfil_con_riesgo.show()

# Al terminar, es buena práctica cerrar la sesión para liberar memoria
spark.stop()
print("Sesión de Spark finalizada.")
```
