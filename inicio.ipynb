```python
import os
import sys

# 1. Rutas de Java y Hadoop
os.environ["JAVA_HOME"] = r"C:\java\jdk-17.0.19+10" 
os.environ["HADOOP_HOME"] = r"C:\hadoop"

# =========================================================================
# SOLUCIÓN AL MISMATCH: Forzar a Spark a usar el mismo ejecutable de Python 3.11
# =========================================================================
ruta_python_311 = r"C:\Users\Luis Alberto\AppData\Local\Programs\Python\Python311\python.exe"
os.environ["PYSPARK_PYTHON"] = ruta_python_311
os.environ["PYSPARK_DRIVER_PYTHON"] = ruta_python_311
# =========================================================================

# Mantener el resto de tu configuración idéntica
sys.path.append(os.environ["HADOOP_HOME"] + "\\bin")
os.environ["PATH"] += os.pathsep + os.path.join(os.environ["HADOOP_HOME"], "bin")
os.environ["SPARK_LOCAL_IP"] = "127.0.0.1"

ruta_librerias = r"C:\Users\Luis Alberto\AppData\Local\Programs\Python\Python311\Lib\site-packages"
if ruta_librerias not in sys.path:
    sys.path.insert(0, ruta_librerias)

# Inicializar Spark
from pyspark.sql import SparkSession

print("--> Lanzando SparkSession alineada con Python 3.11...")
spark = SparkSession.builder \
    .appName("ProyectoAML_Local") \
    .master("local[1]") \
    .config("spark.driver.host", "127.0.0.1") \
    .config("spark.driver.bindAddress", "127.0.0.1") \
    .config("spark.ui.enabled", "false") \
    .config("spark.driver.extraJavaOptions", "-Djava.net.preferIPv4Stack=true") \
    .config("spark.executor.extraJavaOptions", "-Djava.net.preferIPv4Stack=true") \
    .getOrCreate()

# Prueba de fuego final
data = [
    ("TX1001", "Cliente_A", 15000.00),
    ("TX1002", "Cliente_B", 2500.50)
]
df = spark.createDataFrame(data, ["ID", "Cliente", "Monto"])

print("\n" + "="*60)
print("   ¡ÉXITO TOTAL! ENTORNO CONSOLIDADO Y CORRIENDO")
print("="*60)
df.show()
```
